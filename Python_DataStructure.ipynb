4)Describe how dictionaries store data

  ->Dictionaries in Python store data as key-value pairs. Each key is unique and associated with a specific value. This structure allows for efficient retrieval of values based on their corresponding keys.

  -Key-Value Pairs: The fundamental unit of storage in a dictionary is the key-value pair. The key acts as a unique identifier for the value. Keys must be immutable data types (like strings, numbers, or tuples containing only immutable elements), while values can be of any data type (mutable or immutable).

  -Hashing: Dictionaries use a hash table (also known as a hash map) internally to store key-value pairs. When you add a key-value pair to a dictionary, Python calculates a hash value for the key. This hash value is used to determine where in the hash table the key-value pair should be stored.

  -Hash Table Structure: A hash table is essentially an array of buckets or slots. The hash value of a key is used to map the key to a specific bucket in the hash table.

  -Collision Handling: It's possible for different keys to have the same hash value. This is called a hash collision. When a collision occurs, dictionaries use various techniques to handle it, such as:

  -Chaining: Storing multiple key-value pairs with the same hash value in a linked list within that bucket.

  -Open Addressing: Probing for the next available slot in the hash table if the calculated bucket is already occupied.
Efficient Lookups: When you try to access a value using its key, Python calculates the hash value of the key again and uses it to go directly to the corresponding bucket in the hash table. This makes dictionary lookups very efficient, with an average time complexity of O(1) (constant time), regardless of the size of the dictionary.

5)Why might you use a set instead of a list in Python

   ->You might choose to use a set instead of a list in Python for several reasons, primarily due to the inherent properties of sets that make them more suitable for certain tasks.

  Here are the main reasons why you might prefer a set over a list:

  Uniqueness: Sets automatically enforce uniqueness. They cannot contain duplicate elements. If you add a duplicate element to a set, it will be ignored. This is very useful when you need to work with a collection of distinct items and want to avoid duplicates without manually checking for them. Lists, on the other hand, can contain duplicate elements.

  my_list = [1, 2, 2, 3, 3, 3]
  my_set = set(my_list)
  print(my_list) # Output: [1, 2, 2, 3, 3, 3]
  print(my_set)  # Output: {1, 2, 3}
  Membership Testing: Checking if an element is present in a set is significantly faster than checking for membership in a list, especially for large collections. This is because sets are implemented using hash tables, which provide average O(1) (constant time) complexity for membership testing. Lists, on the other hand, require iterating through the elements, resulting in O(n) (linear time) complexity in the worst case.

 import time

 large_list = list(range(1000000))
 large_set = set(large_list)

 start_time = time.time()
 print(999999 in large_list)
 end_time = time.time()
 print(f"List membership test time: {end_time - start_time} seconds")

 start_time = time.time()
 print(999999 in large_set)
 end_time = time.time()
 print(f"Set membership test time: {end_time - start_time} seconds")
 Mathematical Set Operations: Sets provide built-in support for mathematical set operations like union, intersection, difference, and symmetric difference. These operations are very efficient when performed on sets. While you can perform similar operations with lists, it would typically require more complex and less efficient code.

 set1 = {1, 2, 3, 4}
 set2 = {3, 4, 5, 6}

print(set1.union(set2))         # Output: {1, 2, 3, 4, 5, 6}
print(set1.intersection(set2))  # Output: {3, 4}
print(set1.difference(set2))    # Output: {1, 2}
print(set1.symmetric_difference(set2)) # Output: {1, 2, 5, 6}
Removing Duplicates: As mentioned earlier, sets are a very efficient way to remove duplicate elements from a collection.

7)How do tuples ensure data integrity in Python

   ->1. Immutability = Protection Against Accidental Changes

If data shouldn’t change (like configuration values, coordinates, database records), a tuple guarantees that nobody will mess it up during runtime.

2. Safer for Sensitive Data
When you pass a tuple to a function, you can be 100% sure the data will not be modified inside that function.

3. Hashable & Usable as Dictionary Keys
Only immutable objects can be hashable.

4. Tuples Prevent Logical Errors
If your code relies on values never changing (like matrix dimensions, GPS coordinates, RGB values), tuples enforce that rule.

5. Better for Multi-Threading
Because they can’t change, tuples are thread-safe.
Multiple threads can use the same tuple without ruining each other’s data.
Lists can get corrupted under multi-thread access.

6. They Communicate Intent
When a developer sees a tuple, it signals:
“Don’t change this data.”
That’s clarity + integrity in your program design

  ->A hash table is a data structure that stores key–value pairs and allows very fast lookup, insertion, and deletion — typically in O(1) average time.

	1.	You give it a key.

	4.	The value is stored at that index.
In Python, dictionaries (dict) are implemented using hash tables under the hood.

	•	hashes "name"
